```python
#seed for reproducibility
import torch

def fix_torch_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

fix_torch_seed()
```

## general pretrained model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("upstage/TinySolar-248m-4k")
model = AutoModelForCausalLM.from_pretrained("upstage/TinySolar-248m-4k")
```

```python
prompt = "I am an engineer. I love"
```

```python
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
from transformers import TextStreamer
streamer = TextStreamer(
    tokenizer,
    skip_prompt=True, # If you set to false, the model will first return the prompt and then the generated text
    skip_special_tokens=True
)
```

```python
outputs = model.generate(
    **inputs, 
    streamer=streamer, 
    use_cache=True,
    max_new_tokens=128,
    pad_token_id=tokenizer.eos_token_id,
    do_sample=False, 
    temperature=0.0,
    repetition_penalty=1.1
)
```

## Python samples with pretrained Python model


```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("upstage/TinySolar-248m-4k-py")
model = AutoModelForCausalLM.from_pretrained("upstage/TinySolar-248m-4k-py")
```

```python
prompt =  "def find_max(numbers):"

inputs = tokenizer(
    prompt, return_tensors="pt"
).to(model.device)

streamer = TextStreamer(
    tokenizer,
    skip_prompt=True,
    skip_special_tokens=True
)

outputs = model.generate(
    **inputs,
    streamer=streamer,
    pad_token_id=tokenizer.eos_token_id,
    use_cache=True,
    max_new_tokens=128,
    do_sample=False,
    temperature=0.0,
    repetition_penalty=1.1
)
```

```python
def find_max(numbers):
   max = 0
   for num in numbers:
       if num > max:
           max = num
   return max
```

```python
find_max([1,3,5,1,6,7,2])
```
